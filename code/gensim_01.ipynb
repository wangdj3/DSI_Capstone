```python
#!/usr/bin/python3

"""
    opensearch.py
    MediaWiki API Demos
    Demo of `Opensearch` module: Search the wiki and obtain
	results in an OpenSearch (http://www.opensearch.org) format
    MIT License
"""

import requests

S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"
searchterm = 'game of thrones (TV series)'

PARAMS = {
    "action": "opensearch",
    "namespace": "0",
    "search": searchterm,
    "limit": "5",
    "format": "json"
}

R = S.get(url=URL, params=PARAMS)
DATA = R.json()

print(DATA)
```

```python
# Reference: https://www.jcchouinard.com/wikipedia-api/
# Reference: https://www.mediawiki.org/wiki/API:Parsing_wikitext
# Reference: https://www.mediawiki.org/wiki/API:Get_the_contents_of_a_page

import requests
 
subject = 'hawkeye (TV series)'
url = 'https://en.wikipedia.org/w/api.php'
params = {
        'action': 'query',
        'format': 'json',
        'titles': subject,
        'prop': 'extracts',
        # 'exintro': True,
        'explaintext': True,
        'redirects':True
    }
 
response = requests.get(url, params=params)
data = response.json()
 
page = next(iter(data['query']['pages'].values()))
print(page['extract'])
```

```python
# Reference: https://www.mediawiki.org/wiki/API:Parsing_wikitext
#            https://www.mediawiki.org/wiki/API:Get_the_contents_of_a_page

#!/usr/bin/python3

"""
    parse.py

    MediaWiki API Demos
    Demo of `Parse` module: Parse content of a page

    MIT License
"""

import requests

S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

PARAMS = {
    "action": "parse",
    "page": "Pet door",
    "format": "json"
}

R = S.get(url=URL, params=PARAMS)
DATA = R.json()

print(DATA["parse"]["text"]["*"])
```

```python
soup = BeautifulSoup(DATA["parse"]["text"]["*"], "html.parser")
```
