#### Read in IMDB data

#### Read in TheMovieDB data

```python
tmdb_articles_df = pd.read_csv(f'{data_directory}tmdb_articles_df.csv')
tmdb_articles_df.head()
```

```python
tmdb_articles_df.drop(columns='imdb_href', inplace=True)
```

#### Merge dataframes

```python
# merged_df = pd.concat([no_dupes, wiki_articles_df, tmdb_articles_df], axis=1)
merged_df = pd.concat([no_dupes, tmdb_articles_df], axis=1)
merged_df.shape
```

```python
merged_df.head()
```

```python
merged_df.columns
```

```python
merged_df.fillna("", inplace=True)
```

#### Creating a 'is_anime' column

```python
# merged_df[merged_df['imdb_genre_tags'].str.contains("Animation")] # [c] 481

# merged_df[merged_df['tmdb_keywords'].str.contains("anime")] # [a] 191
# merged_df[merged_df['tmdb_keywords'].str.contains("anime") & merged_df['imdb_genre_tags'].str.contains("Animation")] # [a & c] 190
# merged_df[merged_df['tmdb_keywords'].str.contains("anime") & ~merged_df['imdb_genre_tags'].str.contains("Animation")] #[a & ~c] 1 - live-action Cowboy Bebop

# merged_df[merged_df['tmdb_keywords'].str.contains("anime") & merged_df['origin_country'].str.contains("JP")] # [a & b] 180
# merged_df[merged_df['tmdb_keywords'].str.contains("anime") & ~merged_df['origin_country'].str.contains("JP")] #[a & ~b] 11 - Western cartoons in anime style

# merged_df[merged_df['origin_country'].str.contains("JP")] # [b] 184
# merged_df[merged_df['origin_country'].str.contains("JP") & merged_df['imdb_genre_tags'].str.contains("Animation")] # [b & c] 182
# merged_df[merged_df['origin_country'].str.contains("JP")
#          & merged_df['imdb_genre_tags'].str.contains("Animation")
#          & ~merged_df['tmdb_keywords'].str.contains("anime")
#         ] # 2 - Yu-Gi-Oh & Battle Angel - [~a & b & c] both should be anime

# merged_df['wiki_text'].str.contains("anime")  # [d] 261
# merged_df['tmdb_genres'].str.contains("Animation") # [e] 465

a = merged_df['tmdb_keywords'].str.contains("anime") #191
b = merged_df['origin_country'].str.contains("JP") #184
c = merged_df['imdb_genre_tags'].str.contains("Animation") #481
# d = merged_df['wiki_text'].str.contains("anime")  #261
# e = merged_df['tmdb_genres'].str.contains("Animation") #465

# merged_df[a & c & ~b]
```

```python
merged_df['is_anime'] = 0
merged_df.loc[a&c, 'is_anime'] = 1
```

#### Write to disk

```python
merged_df.to_csv(f'{data_directory}merged_nowiki_df.csv', index=False)
```

```python
merged_df.to_json(f'{data_directory}merged_nowiki_df.json')
```
